## HISTORICAL DATA OF CANOO USING YAHOO FINANCE

### Before starting we need install a library called yfinance to download the data available 

```python
import yfinance as yf
```

```python
canoo = yf.Ticker("GOEV")
```

```python
canoo.info #stock information
```

```python
history = canoo.history(period="1mo")
history
```

```python
canoo.history_metadata
```

```python
canoo.get_shares_full(start="2024-02-19",end=None)
```
